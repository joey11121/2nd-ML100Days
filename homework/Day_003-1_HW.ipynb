```python
data = {'國家': ['Taiwan', 'Japan', 'United States', 'South Korea', 'Singapore', 'China', 'United Kingdom', 'North Korea',
               'Germany', 'Denmark', 'Canada', 'Brazil', 'Argentina', 'Portugal', 'Spain', 'Malaysia', 'Phillipines',
               'Egypt', 'Italy', 'Greece', 'Czech', 'Astonia', 'Sweden', 'Finland', 'Norway', 'France', 'Russia', 'South Africa', 
              'Mexico', 'Saudi Arab'] ,
        '人口': np.random.randint(10000, 1000000000, size = 30)}
country_pop = pd.DataFrame(data)
country_pop.sort_values(by = ['人口'],ascending = [False])  
```
